### Inisialisasi urllib

```python
import urllib

r = urllib.request.urlopen('https://www.liputan6.com/bisnis').read()
print(type(r))
print(str(r[:2000])+'...')
print(type(r))
```

### Inisialisasi BeautifulSoup

```python
from bs4 import BeautifulSoup as bs

soup = bs(r, "lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

### Mendapatkan Semua Link

```python
links = soup.find_all('a')
print(len(links))
    
for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
berita_link = [link for link in links if '/read/' in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

### Mendapatkan Isi Berita

```python
html = urllib.request.urlopen("https://www.liputan6.com/bisnis/read/3676495/bei-ubah-mekanisme-pre-closing-pada-2019").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("div", {"class": "read-page-upper"}).find("h1", {"class": "read-page--header--title entry-title"}).text.strip()
isi_berita = soup.find("div", {"class": "article-content-body__item-content"}).text
tanggal = soup.find("div",{"class": "read-page--header--author__wrapper"}).find("time", {"class": "read-page--header--author__datetime updated"}).text

#print(tanggal.strip())
tanggal = tanggal.strip().split(',')[0]
print(tanggal)
print(isi_berita)
print(judul)
```
