```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## JOIN

### INNER JOIN

```python
run("SELECT productCode, productName, textDescription, t1.productline, t2.productline FROM products t1 INNER JOIN productlines t2 ON t1.productline = t2.productline")

#digunakan jika kolom yang dapat join memiliki nama kolom yang berbeda
```

```python
run("SELECT productCode, productName, textDescription, t1.productline, t2.productline FROM products t1 INNER JOIN productlines t2  USING (productline)")
```

```python
run("""SELECT T1.orderNumber, status, SUM(quantityOrdered * priceEach) total FROM orders T1 INNER JOIN orderdetails T2
        ON T1.ordernumber = t2.ordernumber
        GROUP BY ordernumber""")
```

```python
run("""SELECT orderNumber, productName, msrp, priceEach FROM products t1 INNER JOIN orderdetails t2
    ON t1.productcode = t2.productcode AND t1.msrp > t2.priceEach AND t1.productcode = 's10_1678'""")
```

### LEFT JOIN

```python
#Left Join, based on tabel 1 (misal ada tabel 1 dan 2)

run(""" SELECT t1.customerNumber, customerName, ordernumber, status 
    FROM customers t1 LEFT JOIN orders t2 
    ON t1.customernumber = t2.customernumber """)
```

```python
run(""" SELECT t1.customerNumber, customerName, ordernumber, status 
    FROM customers t1 LEFT JOIN orders t2 
    ON t1.customernumber = t2.customernumber
    WHERE ordernumber is NULL""")
```

```python
run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails USING (orderNumber)
WHERE
    orderNumber = 10123
""")
```

```python
run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails d ON o.orderNumber = d.orderNumber
        AND o.orderNumber = '10123'
""")
```
