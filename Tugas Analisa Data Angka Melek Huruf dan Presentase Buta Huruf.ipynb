# Tugas Analisa Data Angka Melek Huruf dan Presentase Buta Huruf

## Data Angka Melek Huruf

Dataset pertama yang digunakan adalah dataset "Angka Melek Huruf Usia 15 - 24 per Provinsi" dari data.go.id. Angka Melek Huruf (selanjutnya disingkat AMH) merupakan proporsi penduduk usia 15 tahun ke atas yang mempunyai kemampuan membaca dan menulis huruf latin dan huruf lainnya, tanpa harus mengerti apa yang di baca/ditulisnya terhadap penduduk usia 15 tahun ke atas. Dataset ini berisi data AMH seluruh provinsi dari tahun 2005 hingga 2009. Data juga disajikan hingga level kabupaten dan kota. Nilai AMH disajikan dalam angka dengan satuan Persen.

```python
# import pandas
import pandas as pd
```

```python
data_amh = pd.read_csv('processed-angka-melek-huruf-usia-15-24-2005-2009.csv', delimiter = ',', encoding = "utf-8")
```

```python
data_amh #dataset "Angka Melek Huruf Usia 15 - 24 per Provinsi" 
```

```python
data_amh.columns
```

Data terdiri dari 2410 baris serta enam kolom yang masing-masing variabel berupa kode provinsi, nama provinsi, kode kabupaten/kota, nama kabupaten/kota, tahun, dan AMH.

## Data Presentase Buta Huruf

Dataset yang kedua adalah dataset "Presentase Buta Huruf" dari data.go.id. Presentase buta huruf merupakanProporsi penduduk usia 15 tahun ke atas yang tidak mempunyai kemampuan membaca dan menulis huruf latin dan huruf lainnya terhadap penduduk usia 15 tahun ke atas.  Dataset ini berisi persentase penduduk buta huruf menurut kelompok umur per propinsi dan kelompok umur.  kelompok umur yang dimaksud terdiri dari kelompok usia lima belas tahun ke atas (15+), usia antara lima belas hingga empat puluh empat tahun (15-44), dan usia 45 tahun ke atas (45+). Tidak seperti dataset pertama data ini hanya disajikan pada level provinsi.

```python
data_abh = pd.read_csv('processeddatabutahuruf20032012perpropinsi.csv', delimiter = ',', encoding = "utf-8")
```

```python
data_abh # dataset "Presentase Buta Huruf"
```

```python
data_abh.columns
```

Data terdiri dari 990 baris serta 5 kolom yang masing-masing variabel berupa kode provinsi, nama provinsi, kelompok usia, tahun, dan presentase buta huruf.

## Menyortir Dataset

Agar kedua data set dapat dicari keterkaitannya maka dalam tugas kali ini terdapat beberapa catatan dalam melakukan penyortiran:
1. Untuk dataset "Angka Melek Huruf Usia 15 - 24 per Provinsi" ditentukan nilai AMH provinsi dengan merata-ratakan jumlah nilai AMH kabupaten/kota pada untuk setiap provinsi pada setiap tahun. (Dataset kedua hanya tersedia dalam sajian tingkat provinsi)
2. Kelompok usia yang digunakan pada Dataset "Presentase Buta Huruf" adalah usia 15-44 untuk mendekati rentang usia pada Dataset pertama yaitu 15-24 tahun.
3. Kemudian dikerucutkan kembali hanya data pada tahun 2005 hingga 2009 yang digunakan untuk dataset "Presentase Buta Huruf".

```python
data_amh_prov = data_amh.groupby(['nama_provinsi', 'tahun']).mean()
```

```python
data_amh_prov # dataset "Angka Melek Huruf Usia 15 - 24 per Provinsi" rata rata setiap provinsi untuk setiap tahun
```

```python
data_abh_1544 = data_abh[data_abh['kelompok_usia']=='15-44']
```

```python
# data_abh_1544
```

```python
data_abh_2010 = data_abh_1544[data_abh_1544['tahun'] < 2010] 
```

```python
# data_abh_2010
```

```python
data_abh_baru = data_abh_2010[data_abh_2010['tahun'] > 2004 ]
```

```python
data_abh_baru #Kemudian dikerucutkan kembali hanya data pada tahun 2005 hingga 2009 yang digunakan untuk dataset "Presentase Buta Huruf".
```

```python
data_abh_prov = data_abh_baru.groupby(['nama_provinsi', 'tahun']).mean()
```

```python
data_abh_prov
```

Untuk mengetahui trend dari tahun ke tahun saya mengambil rata-rata nasional yang kemudian disajikan dalam bentuk grafik baik untuk data angka melek huruf ataupun data presentase buta huruf.

```python
#Angka Melek Huruf
ig = data_amh_prov.groupby(['tahun']).mean()['angka_melek_huruf']
ig = ig.plot(kind='line', title ="Rata-Rata Nasional AMH", figsize=(10,5), legend=True, fontsize=12)
ig.set_xlabel("Tahun", fontsize=12)
ig.set_ylabel("AMH", fontsize=12)
```

```python
# Presentase Buta Huruf

ig_abh = data_abh_prov.groupby(['tahun']).mean()['persentase_buta_huruf']
ig_abh = ig_abh.plot(kind='line', title ="Rata-Rata Nasional ABH / Presentase Buta Huruf", figsize=(10,5), legend=True, fontsize=12)
ig_abh.set_xlabel("Tahun", fontsize=12)
ig_abh.set_ylabel("ABH", fontsize=12)
```
