```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine

host = 'localhost'
port = '3306'
username = 'root'
password = ''
database = 'classicmodels'

#create connection to database

engine = create_engine('mysql+pymysql://'+username+':'+password+'@'+host+':'+port+'/'+database)
# '''engine = create_engine('mysql+pymysql://root: @localhost:3306/academic')'''

def run(sql):
    df = pd.read_sql_query(sql, engine)
    return df
```

```python
run("""SELECT
            *
        FROM
            customers
""")
```

```python
#csv to dataframe to mysql
df_fromcsv = pd.read_csv('personalcustomers.csv')
df_fromcsv.to_sql(name = 'personalcustomers', con = engine, if_exists = 'replace', index = False)
#if_exist = 'fail' | 'replace' | 'append'
```

```python
#dataframe to csv/excel

df_customers = run("""SELECT
            *
        FROM
            customers
""")

df_customers.to_excel('customers.xlsx', index = False)
```

```python
#Select all mahasiswa data using run function (pandas)
#data in dataframe type
df_result = run("SELECT * FROM personalcustomers")
print (df_result)
```
